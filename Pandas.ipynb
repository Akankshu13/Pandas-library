# Panda Series

```python
import pandas as pd
import numpy as np
```

```python
#creating an array

a=np.arange(1,7)
print(a)
```

```python
#coverting it into series
a=pd.Series(np.arange(1,7))
print(a)             #we did not give the index, so the index order based on the given data 
```

```python
a=pd.Series(np.arange(1,7),dtype=np.int64)     #here we change the data type
print(a)
```

```python
#if we want to change the index
a=pd.Series(np.arange(1,7),dtype=np.int64,index=['a','b','c','d','e','f'])     #here we change the data type
print(a)
```

```python
c=pd.Series(65,index=[1,2,3,4,5,6])
print(c)
```

```python
d=np.array(['Ram','Shyam','Sita','Gita'])
print(d)
print(d.ndim)     #we use ndim for check the dimension
print()
e=pd.Series(np.array(['Ram','Shyam','Sita','Gita']))
print(e)
```

### we can also do indexing and slicing

```python
b=pd.Series(np.arange(1,10),index=['row1','row2','row3','row4','row5','row6','row7','row8','row9'])
print(b)
```

```python
b['row6':'row9']    #retrieve records by label index: 0 to n
```

```python
b['row1':'row3']
```

```python
b[0:4]    # position based indexing : 0 to n-1 5 
```

```python
b[7:9]
```

```python
b['row9']   #retrieve records by numeric index
```

```python
b[7]
```

```python
x = pd.Series([1,2,3],index = ['a','b','c'])  
                                                     #retrieve the last element  
print(x)
print("\nAccessing index _ value = " , x['c'])       #\n for changing the line
```
